To predict salary the most appropriate approach would be a regression model.
Here instead we just want to estimate which factors (like location, job title, job level, industry sector) lead to high or low salary and work with a classification model. Notice, each job listing is underneath a `div` tag with a class name of `result`. We can use BeautifulSoup to extract those.

## 2.2 Function to scrape job data

Below I created the function for the scraping of the data from the indeed website. As input the URL for the website is needed. Below i will describe the steps of the function: 

1. It will use the url which has been inputed to open a new chrome window of the website. 
2. The code will accept the cookies if needed. This was a problem which was coming up on the Europe indeed websites. 
3. A loop will iterate through the following steps:

   3.1 If available the following information will be collected from the jobs listed on the website: 
            - title 
            - location
            - company name
            - salary 
            - salary estimation 
            - link to the specific job offer
   3.2 The information will be saved in a dataframe. 
   
   3.3 A VPN counter has been included to check if the VPN location should be adjusted. From experience I after every 3rd page a change would be needed.
   
   3.4 At the end the code is clicking to button for the next page. As there are sometimes more than 20 pages and sometimes only 3 the code is trying different buttons to move on. If there is no button for the next page available anymore it will break the loop and end.
   
   3.5 After the end of every iteration the code is saving the data into a csv file. Like this I prevent loosing progress if the code or the machine are crashing. 
   
   3.6 Depending on step 3.4 the code will either start again on section 3.1. or break. 

4. After the loop ended the code will print out the link used for the scraping. Like this I can keep track which URLs have already been completed. 
5. The code will close the browser window after all URLs have been scrapped . 
    

```python
def scraping (url):

    global df_us2 # this variable needs to be adjusted as it is related to the dataframe
    
    # open a new Chrome page for scraping
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(url)
    time.sleep(10)

    # accept the cookies if needed: 
    try: driver.find_elements_by_xpath('//*[@id="onetrust-accept-btn-handler"]')[0].click()
    except:
        None

    
    driver.get(url)
    
    #setting of the variables for the loop
    jobs = []
    vpn_count = 0
    page_count = 0
    
    driver.implicitly_wait(20)
    
    for n in range(50):
        # I set the range to 50 as I would like to scrape as many pages as possible.
        # If there less than 50 pages the break function will end the loop.
        for job in driver.find_elements_by_class_name('result'):
            # the loop will iterate through the job search results on the page 
            soup = BeautifulSoup(job.get_attribute('innerHTML'),'html.parser')
            # beatifull soup will enter into the attributes of the jobs to extract them
        
            try:
                title = soup.find(class_="jobTitle").text # job itle
            except:
                title = 'None'
            try:
                location = soup.find(class_="companyLocation").text # company location 
            except:
                location = 'None'
            try:
                company = soup.find(class_="companyName").text.replace("\n","").strip() # company name 
            except:
                company = 'None'
            try:
                salary = soup.find(class_="salary-snippet-container").text # salary 
            except: 
                salary = "None"
            try:
                salary_estimation = soup.find(class_="estimated-salary").text.replace("Estimated","").strip() #estimated salary
            except:
                salary_estimation = 'None'  
            


            sum_div = job.find_element_by_class_name('job_seen_beacon')
            try:    
                sum_div.click()
            except:
                close_button = driver.find_elements_by_class_name('popover-x-button-close')[0]
                close_button.click()
                sum_div.click()            
            driver.implicitly_wait(1)
            
            # collecting the links of the specific jobs
            
            liens = job.find_element_by_tag_name("a")

            links = liens.get_attribute("href")
            
            # sleep timer to avoid captures
            time.sleep(1)

            # attached the collected data into the dataframe, name of dataframe needs to be adjusted 
            df_us2 = df_us2.append({'Title':title,
                                            'Location':location,
                                            "Company":company, 
                                            "Salary":salary,
                                            "Salary Estimation":salary_estimation,
                                            "Links":links},
                                           ignore_index=True)
            Links_list = df_us2['Links'].tolist()
        
        
        #update VPNcount and if needed change of location 
        #every three pages we will change the location
        vpn_count +=1
        if vpn_count % 3 == 0:
            nord_vpn_update()
           # we will give the VPN 10 seconds to change the location
            time.sleep(10)   
    
        # move to next page
        if page_count ==0:
            try:
                driver.find_elements_by_xpath('//*[@id="resultsCol"]/nav/div/ul/li[6]/a/span')[0].click()
            except:
                driver.find_elements_by_xpath('//*[@id="resultsCol"]/nav/div/ul/li[2]/a/span')[0].click()
            page_count+= 1
            #sleep timer to avoid captures
            time.sleep(5)
        else:
            try:
                driver.find_elements_by_xpath('//*[@id="resultsCol"]/nav/div/ul/li[7]/a/span')[0].click()
                time.sleep(5)
            except:
                break
                
        # save the data to a csv, name of csv needs to be adjusted 
        df_us2.to_csv(f'df_us2.csv')
        
    # print urls to track in case the code breaks
    print(url)
    print(64*"-")
    
    # close the browser window
    driver.close()
    nord_vpn_update()
    time.sleep(5)
```

### 4.1.4 Creating final salary Column

Within a next step I will convert the monthly etc. salaries into annual salaries. 

As we have two salaries column we will first use the salary from the companies and if nothing had been provided the salary estimation. 

For the daily, hour and weekly salary we will assume that there are 260 working days within a year. We deduct 30 days of Holidays and estimate 8 hours per day working.

```python
def salary_conv(row):
    row["salary_converted"] = np.nan
    try:
        if row["salary"] >0:
                if row["salary_period"] == "year":
                    row["salary_converted"] = round(row["salary"],2)
                elif row["salary_period"] == "month":
                    row["salary_converted"] = round(row["salary"]*12,2)
                elif row["salary_period"] == "week":
                    row["salary_converted"] = round(row["salary"]*230/7,2)
                elif row["salary_period"] == "day":
                    row["salary_converted"] = round(row["salary"]*230,2)
                elif row["salary_period"] == "hour":
                    row["salary_converted"] = round(row["salary"]*8*230)
        else:  
            if row["salary_estimation_period"] == "year":
                row["salary_converted"] = round(row["salary_estimation"])
            elif row["salary_estimation_period"] == "month":
                row["salary_converted"] = round(row["salary_estimation"]*12)
            elif row["salary_estimation_period"] == "week":
                row["salary_converted"] = round(row["salary_estimation"]*230/7)
            elif row["salary_estimation_period"] == "day":
                row["salary_converted"] = round(row["salary_estimation"]*230)
            elif row["salary_estimation_period"] == "hour":
                row["salary_converted"] = round(row["salary_estimation"]*230)
        return row["salary_converted"]
    except:
        pass
```

```python
df_jobs["salary_converted"] = df_jobs.apply(lambda row: salary_conv(row), axis=1)
```

First i would like to describe the evaluation measurements:

**Accuracy** : This is the most naive and commonly used metric in the context of classification. It is just the mean of correct predictions.

accuracy = (Total correct prediction)/Total prediction

= (TP + TN)/( TP + TN + FP + FN)

**<u>Precision and Recall</u>**

- Precision : It is defined as proportion of correctly predicted positive outcome among all prediction. In other words of all the predicted positive outcome how many of them are actually positive. Precision = TP /(TP + FP)
- Recall : It is defined as proportion of correctly predicting positive outcome among all actual positive. In other words of all the actual positive outcome how many of them we have been able to predict as positive. Recall = TP/(TP + FN)

Taking into consideration that our boss wants rather tell a client incorrectly that they would get a lower salary job than tell a client incorrectly that they would get a high salary job. The precision for the low salary section should be as high as possible. 


**<u>Receiver Operating Characteristic (ROC)</u>**

A ROC curve is constructed by plotting the true positive rate (TPR) against the false positive rate (FPR). The true positive rate is the proportion of observations that were correctly predicted to be positive out of all positive observations (TP/(TP + FN)). Similarly, the false positive rate is the proportion of observations that are incorrectly predicted to be positive out of all negative observations (FP/(TN + FP)). For example, in medical testing, the true positive rate is the rate in which people are correctly identified to test positive for the disease in question. A discrete classifier that returns only the predicted class gives a single point on the ROC space. But for probabilistic classifiers, which give a probability or score that reflects the degree to which an instance belongs to one class rather than another, we can create a curve by varying the threshold for the score. Note that many discrete classifiers can be converted to a scoring classifier by looking inside their instance statistics. For example, a decision tree determines the class of a leaf node from the proportion of instances at the node. 

<u>Interpreting the ROC curve</u>

The ROC curve shows the trade-off between sensitivity (or TPR) and specificity (1  FPR). Classifiers that give curves closer to the top-left corner indicate a better performance. As a baseline, a random classifier is expected to give points lying along the diagonal (FPR = TPR). The closer the curve comes to the 45-degree diagonal of the ROC space, the less accurate the test.

Note that the ROC does not depend on the class distribution. This makes it useful for evaluating classifiers predicting rare events such as diseases or disasters. In contrast, evaluating performance using accuracy (TP + TN)/(TP + TN + FN + FP) would favor classifiers that always predict a negative outcome for rare events.

ROC curves are very good for balanced classifications. Which is the case here for us. 
    

**<u>Area under curve (AUC)</u>**

To compare different classifiers, it can be useful to summarize the performance of each classifier into a single measure. One common approach is to calculate the area under the ROC curve, which is abbreviated to AUC. It is equivalent to the probability that a randomly chosen positive instance is ranked higher than a randomly chosen negative instance, i.e. it is equivalent to the two sample Wilcoxon rank-sum statistic.

A classifier with high AUC can occassionally score worse in a specific region than another classifier with lower AUC. But in practice, the AUC performs well as a general measure of predictive accuracy. The AUC range is between 0 and 1.

For our valuation this means we would like to have a high AUC for the class 1, higher salary.

Within this jupyter notebook I prepared a classification model for high and low salaries which have been extracted from the website of Indeed. Within the following writeup I will explain the procedure and highlight risks and limitations.

To collect enough data I focused on different cities: 'London', 'Manchester', 'Birmingham', 'Glasgow', 'Liverpool', Hamburg", Berlin", "Frankfurt", "Paris", "Amsterdam", 'New York', 'Chicago', 'San Francisco', 'Austin', 'Seattle', 'Los Angeles', 'Philadelphia', 'Atlanta', 'Dallas', 'Pittsburgh', 'Portland', 'Phoenix', 'Denver', 'Houston', 'Miami', "Luxembourg", "Cambridge". I decided to extract the Job title, company name, location of the job and the link of the different jobs. 
While extracting the data I got blocked several times by the captures from the indeed website. To avoid this I created a short function to change my virtual location with a VPN client. A VPN client is a program which creates a connection to a server within a different country, for example Chile, and simulates that I am accessing the indeed website from Chile. This was a huge step to avoid the captures and my program managed to extract a lot of data.  

After extracting the data I started to clean the different columns. For my later analysis it was important to have the location and salary columns cleaned. 
Within the salary section several different information had been included. For example, was the salary sometimes provided in a range. For these I calculated the average of the range. Also, the salaries have been showed on different time horizons annually, monthly, weekly or even per hour. I decided to calculate the yearly salary for all of them. Also, the different source countries had to been taken into consideration. As the salary data was included in USD, GBP and EUR I transferred the USD and EUR once into GBP to make them comparable. 

The next column to be cleaned was the location column. Here a lot of information had been included which I didnt need for my later analysis. I scanned the column for my main cities (list above) and whenever it was clear that a row could be allocated to one of these cities the location has been adjusted to the city name. For some cities like Amsterdam, Paris and Luxembourg I concluded that all close cities will also be allocated to the main city Paris, Amsterdam or Luxembourg.
